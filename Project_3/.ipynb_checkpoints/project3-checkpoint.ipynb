### 1. Meet Dr. Ignaz Semmelweis

<p>This is Dr. Ignaz Semmelweis, a Hungarian physician born in 1818 and active at the Vienna General Hospital. If Dr. Semmelweis looks troubled it's probably because he's thinking about <em>childbed fever</em>: A deadly disease affecting women that just have given birth. He is thinking about it because in the early 1840s at the Vienna General Hospital as many as 10% of the women giving birth die from it. He is thinking about it because he knows the cause of childbed fever: It's the contaminated hands of the doctors delivering the babies. And they won't listen to him and <em>wash their hands</em>!</p>
<p>In this notebook, we're going to reanalyze the data that made Semmelweis discover the importance of <em>handwashing</em>. Let's start by looking at the data that made Semmelweis realize that something was wrong with the procedures at Vienna General Hospital.</p>

```python
# Read datasets/yearly_deaths_by_clinic.csv into yearly
yearly_deaths_df = pd.read_csv('datasets/yearly_deaths_by_clinic.csv')

# Print out yearly_deaths_df data-frame
yearly_deaths_df
```

```python
print("Shape of the yearly_deaths_df dataset:", yearly_deaths_df.shape)

print("Data-set column information:")
yearly_deaths_df.info()
```

```python
# Number of unique values in each column
print( yearly_deaths_df.nunique() )

print("Unique values in clinic column:", yearly_deaths_df['clinic'].unique())
```

### 2. The alarming number of deaths
<p>The table above shows the number of women giving birth at the two clinics at the Vienna General Hospital for the years 1841 to 1846. You'll notice that giving birth was very dangerous; an <em>alarming</em> number of women died as the result of childbirth, most of them from childbed fever.</p>
<p>We see this more clearly if we look at the <em>proportion of deaths</em> out of the number of women giving birth. Let's zoom in on the proportion of deaths at Clinic 1.</p>

```python
yearly_deaths_df.head(2)
```

```python
# Calculate proportion of deaths per no. births
yearly_deaths_df['deaths_proportion'] = yearly_deaths_df['deaths'] / yearly_deaths_df['births']

# Extract Clinic 1 data into clinic_1 and Clinic 2 data into clinic_2
clinic_1 = yearly_deaths_df[ yearly_deaths_df['clinic'] == 'clinic 1']
clinic_2 = yearly_deaths_df[ yearly_deaths_df['clinic'] == 'clinic 2']

# Print out clinic_1
display("First Five rows of clinic_1:", clinic_1.head())
print("Shape of clinic_1:", clinic_1.shape)

# Print out clinic_1
display("First Five rows of clinic_2:", clinic_2.head())
print("Shape of clinic_2:", clinic_2.shape)
```

### 3. Death at the clinics
<p>If we now plot the proportion of deaths at both Clinic 1 and Clinic 2  we'll see a curious pattern…</p>

```python
# Plot yearly proportion of deaths at the two clinics

plt.figure(figsize = (10, 6))
plt.plot(clinic_1['year'], clinic_1['deaths_proportion'], '*-', label = 'Clinic 1')
plt.plot(clinic_2['year'], clinic_2['deaths_proportion'], '*-', label = 'Clinic 2')

plt.title("Death Proportion in Different Years in Two Clinics")
plt.xlabel("Year")
plt.ylabel("Death Proportion")

plt.legend()

plt.show()
```

### 4. The handwashing begins
<p>Why is the proportion of deaths consistently so much higher in Clinic 1? Semmelweis saw the same pattern and was puzzled and distressed. The only difference between the clinics was that many medical students served at Clinic 1, while mostly midwife students served at Clinic 2. While the midwives only tended to the women giving birth, the medical students also spent time in the autopsy rooms examining corpses. </p>
<p>Semmelweis started to suspect that something on the corpses spread from the hands of the medical students, caused childbed fever. So in a desperate attempt to stop the high mortality rates, he decreed: <em>Wash your hands!</em> This was an unorthodox and controversial request, nobody in Vienna knew about bacteria at this point in time. </p>
<p>Let's load in monthly data from Clinic 1 to see if the handwashing had any effect.</p>

```python
# Read datasets/monthly_deaths.csv into monthly
monthly_deaths_df = pd.read_csv('datasets/monthly_deaths.csv', parse_dates = ['date'])

display("First 10 rows of monthly_deaths_df:", monthly_deaths_df.head(10))

print("Number of records in monthly_deaths_df:", monthly_deaths_df.shape)

# Do we have missing in the monthly_deaths_df
monthly_deaths_df.isnull().sum()
```

```python
print("Columns Data-type:\n", monthly_deaths_df.dtypes)
```

```python
# Calculate proportion of deaths per no. births
monthly_deaths_df['deaths_proportion'] = monthly_deaths_df['deaths'] / monthly_deaths_df['births']

# Print out the first 05 rows in monthly
monthly_deaths_df.head(5)
```

### 5. The effect of handwashing
<p>With the data loaded we can now look at the proportion of deaths over time. In the plot below we haven't marked where obligatory handwashing started, but it reduced the proportion of deaths to such a degree that you should be able to spot it!</p>

```python
# Plot monthly proportion of deaths
plt.figure(figsize = (20, 6))

plt.plot(monthly_deaths_df['date'], monthly_deaths_df['deaths_proportion'], '*-')

plt.title("Death Proportion Between 1841 and 1849")
plt.xlabel("Date")
plt.ylabel("Monthly Death Proportion")


plt.show()
```

### 6. The effect of handwashing highlighted
<p>Starting from the summer of 1847 the proportion of deaths is drastically reduced and, yes, this was when Semmelweis made handwashing obligatory. </p>
<p>The effect of handwashing is made even more clear if we highlight this in the graph.</p>

```python
# Date when handwashing was made mandatory
handwashing_start = pd.to_datetime('1847-06-01')

# Split monthly into before and after handwashing_start
before_washing = monthly_deaths_df[ monthly_deaths_df['date'] < handwashing_start ]
after_washing = monthly_deaths_df[ monthly_deaths_df['date'] >= handwashing_start ]
```

```python
print("Time Period Before Handwash:\nMinimum date before handwashing obligated:", before_washing['date'].min() )
print("maximum date before handwashing obligated:", before_washing['date'].max(), '\n')

print("Time Period After Handwash:\nMinimum date After handwashing obligated:", after_washing['date'].min() )
print("maximum date before handwashing obligated:", after_washing['date'].max() )
```

```python
display("First Five rows of before_washing:", before_washing.head())
display("First Five rows of after_washing:", after_washing.head())
```

```python
# Plot monthly proportion of deaths before and after handwashing
plt.figure(figsize = (20, 6))

plt.plot(before_washing['date'], before_washing['deaths_proportion'], '*-', label = 'Before Handwashing')
plt.plot(after_washing['date'], after_washing['deaths_proportion'], '*-', label = 'After Handwashing')

plt.title("Monthly Death Proportion Before and After Handwashing")
plt.xlabel("Date")
plt.ylabel("Death Proportion")

plt.legend()

plt.show()
```

### 7. More handwashing, fewer deaths?
<p>Again, the graph shows that handwashing had a huge effect. How much did it reduce the monthly proportion of deaths on average?</p>

```python
# Difference in mean monthly proportion of deaths due to handwashing
before_proportion = before_washing['deaths_proportion']
after_proportion = after_washing['deaths_proportion']

# computing mean (average) of deaths_proportion
before_washing_mean = np.mean( before_washing['deaths_proportion'] )
after_washing_mean = np.mean( after_washing['deaths_proportion'] )

print("Difference between two means:", (before_washing_mean - after_washing_mean)*100.0, "%" )
```

```python
# check deviation before mean value and then plotting.
mean_diff_before_washing = before_washing['deaths_proportion'] -  before_washing_mean
mean_diff_after_washing = after_washing['deaths_proportion'] - after_washing_mean

print("Data_type of mean_diff_before_washing:", type(mean_diff_before_washing) )
display("First 10 values of mean_diff_before_washing:", mean_diff_before_washing[:10] )

print("Data_type of mean_diff_after_washing:", type(mean_diff_after_washing) )
display("First 10 values of mean_diff_after_washing:", mean_diff_after_washing[:10] )
```

```python
plt.figure(figsize = (20, 6))

plt.plot(before_washing['date'], mean_diff_before_washing, '*-', label = 'Before Handwashing')
plt.plot(after_washing['date'], mean_diff_after_washing, '*-', label = 'After Handwashing')

plt.title("Mean Difference Death Proportion Monthly Before and After Handwashing")
plt.xlabel("Date")
plt.ylabel("Mean Difference")

plt.legend()

plt.show()
```

### 8. A Bootstrap analysis of Semmelweis handwashing data
<p>It reduced the proportion of deaths by around 8 percentage points! From 10% on average to just 2% (which is still a high number by modern standards). </p>
<p>To get a feeling for the uncertainty around how much handwashing reduces mortalities we could look at a confidence interval (here calculated using the bootstrap method).</p>

```python
before_proportion = before_washing['deaths_proportion']
after_proportion = after_washing['deaths_proportion']

print("First Five values of before_proportion:\n", before_proportion[:5])
print("First Five values of after_proportion:\n", after_proportion[:5])
```

```python
# A bootstrap analysis of the reduction of deaths due to handwashing
boot_mean_diff = []

# getting 3000 unique sample sets of size 1.0% each
# each sample set will have 1.0% of original poulation
for i in range(3000):
    boot_before = before_proportion.sample(frac = 1, replace = True)
    boot_after = after_proportion.sample(frac = 1, replace = True)
    
    boot_mean_diff.append( np.mean(boot_before) - np.mean(boot_after) )

print("First 10 values of boot_mean_diff:\n", boot_mean_diff[:10])
print("Number of values in boot_mean_diff:", len(boot_mean_diff))
print("Data-type of boot_mean_diff:", type(boot_mean_diff) )
```

```python
print("Minimum Mean Differnce value in boot_mean_diff:", min(boot_mean_diff) )
print("Maximum Mean Difference value in boot_mean_diff:", max(boot_mean_diff) )
```

```python
# Calculating a 95% confidence interval from boot_mean_diff 
confidence_interval = pd.Series( boot_mean_diff ).quantile([0.025, 0.975])
confidence_interval
```

```python
# plotting the histogram plot of Mean differecne of 3000 samples (computed usign Boostraping) 
plt.figure(figsize = (8, 5))

sns.histplot(boot_mean_diff, kde = True)

plt.title("Histogram of Mean Difference of 3000 Samples")
plt.xlabel("Sample Mean Difference")
plt.ylabel("Mean Count")

plt.show()
```

### 9. The fate of Dr. Semmelweis
<p>So handwashing reduced the proportion of deaths by between 6.7 and 10 percentage points, according to a 95% confidence interval. All in all, it would seem that Semmelweis had solid evidence that handwashing was a simple but highly effective procedure that could save many lives.</p>
<p>The tragedy is that, despite the evidence, Semmelweis' theory — that childbed fever was caused by some "substance" (what we today know as <em>bacteria</em>) from autopsy room corpses — was ridiculed by contemporary scientists. The medical community largely rejected his discovery and in 1849 he was forced to leave the Vienna General Hospital for good.</p>
<p>One reason for this was that statistics and statistical arguments were uncommon in medical science in the 1800s. Semmelweis only published his data as long tables of raw data, but he didn't show any graphs nor confidence intervals. If he would have had access to the analysis we've just put together he might have been more successful in getting the Viennese doctors to wash their hands.</p>
